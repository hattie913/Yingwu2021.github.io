#### Data preparation

Preparing the visualizations, we created a consolidated dataset by merging cleaned county labels data, cleaned Bigfoot data, and county data, which contains population and economic data. In order to merge the datasets, we first checked if the data in the ‘state’ and ‘county’ columns are in the same format. For example, we removed some punctuations from the county data. Then, we merged cleaned Bigfoot data and cleaned county label datasets using the ‘fips’ column. We merged again with the formatted county dataset with both ‘state’ and ‘county’ columns. We edited the consolidated dataset by removing duplicated columns and renamed ‘county’ and ‘state’ columns. The consolidated dataset is saved as a ‘merge_bigfoot_continental.pkl’ file.

#### Bigfoot sightings per year

Preparing the visualization of Bigfoot sightings per year, we have decided to use the cleaned Bigfoot dataset, which only contains the data across the contiguous United States. From the dataset, we extracted the data in the date column and stored it in a new dataset. Since the date information includes day, month, and year data, we converted the data from string to datetime format and conducted data extraction to select the year information. The original year data was transformed from two-digit to four-digit format; however, several transformed years are greater than the latest year, which is 2018. We decided to modify the data manually to ensure that the year information is correct and allows us to apply further analysis. If any year is greater than 2018, we subtracted the year by 100. As we are specifically interested in information between 1921 and 2018, we excluded any data that are not within the range. After getting the correct year data, we were able to generate a line graph, which is shown below.

```python
from IPython.display import Image
Image(filename= "ma705 project/image5.png")
```

### Description

The line graph tells us that the number of sightings remained unchanged from 1921 to 1950. Starting from 1950, there was an increase in the number of sightings. There were two peaks around 1978 and 2005, following fluctuations and huge drops in the next few years respectively.
